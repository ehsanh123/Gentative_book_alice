#load data

```python
import numpy as np
from keras.layers import SimpleRNN, Dense, Activation
from keras.models import Sequential

# Download and preprocess text data
import requests

# Download the dataset
url = "https://raw.githubusercontent.com/elboudab/DEL_datasets/main/alice_in_wonderland.txt"
response = requests.get(url)
documents = response.text.lower()

# Remove special characters
documents = documents.replace("\n", " ").replace("\r", " ")
documents = "".join([c if c.isalnum() or c.isspace() else "" for c in documents])

# Characters to index enumerator
chars = sorted(set(documents))
nb_chars = len(chars)
char2index = {c: i for i, c in enumerate(chars)}
index2char = {i: c for i, c in enumerate(chars)}

# Sequence length and step
SEQLEN = 10
STEP = 1

# Create input and output sequences
input_chars = []
label_chars = []

for i in range(0, len(documents) - SEQLEN, STEP):
    input_chars.append(documents[i:i + SEQLEN])
    label_chars.append(documents[i + SEQLEN])

# One-hot encode the characters
#data type has to be NP.bool_
x = np.zeros((len(input_chars), SEQLEN, nb_chars), dtype=np.bool_)
y = np.zeros((len(input_chars), nb_chars), dtype=np.bool_)

for i, input_char in enumerate(input_chars):
    for j, ch in enumerate(input_char):
        x[i, j, char2index[ch]] = 1
    y[i, char2index[label_chars[i]]] = 1
```

#model 1

```python
# Hyperparameters
HIDDEN_UNITS = 128
BATCH_SIZE = 128
N_ITERATIONS = 25
N_EPOCHS_PER_ITERATION = 1
N_PREDICTIONS = 100

# Model architecture
model = Sequential()
model.add(SimpleRNN(HIDDEN_UNITS, return_sequences=False, input_shape=(SEQLEN, nb_chars), unroll=True))
model.add(Dense(nb_chars))
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy", optimizer="rmsprop")

# Training and generating text
for iteration in range(N_ITERATIONS):
    print(f"\nIteration #: {iteration}")
    model.fit(x, y, batch_size=BATCH_SIZE, epochs=N_EPOCHS_PER_ITERATION)

    # Test the model's prediction
    start_index = np.random.randint(0, len(documents) - SEQLEN - 1)
    seed = documents[start_index: start_index + SEQLEN]
    print(f"Seed: {seed}")

    generated = seed
    for _ in range(N_PREDICTIONS):
        x_pred = np.zeros((1, SEQLEN, nb_chars))
        for t, char in enumerate(seed):
            x_pred[0, t, char2index[char]] = 1

        preds = model.predict(x_pred, verbose=0)[0]
        next_index = np.argmax(preds)
        next_char = index2char[next_index]

        generated += next_char
        seed = seed[1:] + next_char

    print(f"Generated Text: {generated}")
```
